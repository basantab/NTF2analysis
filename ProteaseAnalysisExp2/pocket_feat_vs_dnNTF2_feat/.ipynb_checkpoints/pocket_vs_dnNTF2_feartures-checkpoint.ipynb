```python
w_pocket = stable_w_pocket_dn_data.loc[stable_w_pocket_dn_data.nuVolume > 0.0]
sample = w_pocket.sample(n=len(w_pocket['Volume'].values))
colors = ["windows blue", "amber", "red","lime green"]
features = ["h3_len","h2_len","h1_len","has_cHelix","opening",\
            "H3_S3_len","hairpin_len"]
pok_feat = ["Volume","mean_height","mouths","AreaofBiggestMouth","FractionApolarSurfaceArea"]
npok = len(pok_feat)
nfeat = len(features)
fig,ax = plt.subplots(nfeat, npok, figsize=(15,21),squeeze=True)

for k,f in enumerate(features):
    for n,p in enumerate(pok_feat):
        for m,val in enumerate(set(sample[f])):
            bins=np.linspace(min(sample[p]),max(sample[p]),20)
            sns.distplot(sample.loc[sample[f] == val][p],ax=ax[k][n],color=sns.xkcd_palette(colors)[m]\
                         ,kde=False, norm_hist=True,bins=bins,label="%s: %d"%(f,val))
            if n==4: ax[k][n].legend()
            ax[k][n].set_xlabel('')
plt.tight_layout()
plt.show()
```
